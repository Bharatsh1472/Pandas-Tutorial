# Exploring Data

```python
import pandas as pd

data = pd.read_csv("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/customer.csv")
print(data)
print()
# Print the first 10 rows of the DataFrame to get an overview of the data
print(data.head(5))
print()
# Print the last 10 rows of the DataFrame to check the end of the data
print(data.tail(5))
```

```python
# Print the summary information of the DataFrame 'data'
# This includes the number of non-null entries, data types, and memory usage
print(data.info())
```

```python
# Generate descriptive statistics for the DataFrame 'data'
# This includes count, mean, standard deviation, min, max, and quartiles
data.describe()
```

```python
# Print a DataFrame indicating which values are null (True) and which are not (False)
print(data.isnull())
# Print the sum of null values in each column of the DataFrame
print(data.isnull().sum())
```
