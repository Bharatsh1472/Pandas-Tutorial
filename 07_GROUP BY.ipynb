# GROUPBY() METHOD
* In Pandas, GroupBy can be used to group data in many different ways and perform complex data transformations and analysis.
* It is a powerful tool for grouping data and performing operations on these groups.
* The groupby method allows you to split your data into groups based on some criteria, apply a function to each group independently, and then combine the results.

```python
import pandas as pd

data = pd.read_excel("C:/Users/ACER/Desktop/stuff/DS PROJECT/Datasets/ESD.xlsx")
print(data)
```

```python
# Group the data by "Department" and count the number of occurrences of "EEID"
gb_1 = data.groupby("Department").agg({"EEID": "count"})
print(gb_1)
print()

# Group the data by "Job Title" and count the number of occurrences of "EEID"
gb_2 = data.groupby("Job Title").agg({"EEID": "count"})
print(gb_2)
print()

# Group the data by both "Department" and "Gender", then count the number of occurrences of "EEID"
gb_3 = data.groupby(["Department","Gender"]).agg({"EEID": "count"})
print(gb_3)
```

```python
# Grouping the data by 'Country' and calculating the mean age for each country
gb1 = data.groupby(["Country"]).agg({"Age": "mean"})
print(gb1)
print()

# Grouping the data by 'Country' and calculating the mean annual salary for each country
gb2 = data.groupby(["Country"]).agg({"Annual Salary": "mean"})
print(gb2)
print()

# Grouping the data by 'Country' and 'Gender', and calculating the maximum and minimum annual salary for each group
gb3 = data.groupby(["Country","Gender"]).agg({"Annual Salary": ["max", "min"]})
print(gb3)
print()

# Grouping the data by 'Country' and 'Gender', calculating the maximum annual salary and minimum age for each group
gb4 = data.groupby(["Country","Gender"]).agg({"Annual Salary": "max", "Age" : "min"})
print(gb4)
```
